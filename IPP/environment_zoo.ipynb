```python
import os
import matplotlib.pyplot as plt
import dill

# # My modules
# from grid_env import GridEnvironment
# from policy_agent import Policy_Agent
from draw_gridworlds import draw_gridworld, draw_policy, draw_full_policy
# from rl_utils import run_episode, evaluate_agent, get_transition_dict, compute_entropy, get_discount_matrix

plt.rcParams["font.family"] = "Times New Roman"
```

```python
def load_object(filename):    
    with open(filename, 'rb') as f:
        return dill.load(f)
```

```python
for filename in sorted(os.listdir('gridworlds'))[1:-1]:
    
    print(f'=======\nLoading: {filename}')
    env = load_object(f'gridworlds/{filename}')
    
    env.display(display_values=True, dpi=104)
    plt.title(f'"{filename[:-4]}"', y=1, fontsize=14)
    plt.show()
```

```python
plt.figure(figsize=(20,15), dpi=168)
for i, filename in enumerate(sorted(os.listdir('gridworlds'))[1:-1]):
    plt.subplot(3,4,i+1)
    env = load_object(f'gridworlds/{filename}')
    draw_gridworld(env, subplot_mode=True, display_values=True)
    plt.title(f'"{filename[:-4]}"', fontsize=14)
    plt.savefig(f'gridworlds/figs/all.jpeg', bbox_inches='tight')
```
